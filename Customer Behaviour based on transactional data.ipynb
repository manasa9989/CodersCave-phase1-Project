```python
plt.figure(figsize=(16, 8))
grp_counts = data['GRP'].value_counts()
sns.barplot(x=grp_counts.index, y=grp_counts.values)
plt.xticks(rotation=90)
plt.title('Distribution of Purchases by Group')
plt.show()
```

```python
plt.figure(figsize=(26, 18))
subgroup_counts = data['SGRP'].value_counts()
sns.barplot(x=subgroup_counts.index, y=subgroup_counts.values)
plt.xticks(rotation=90, ha='right')  
plt.title('Distribution of Purchases by Subgroup',fontsize=4)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
brand_counts = data['BRD'].value_counts().head(10)
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.xticks(rotation=45)
plt.title('Top 10 Brands by Purchase Count')
plt.show()
```

Here everest is the top most brand 

```python
plt.figure(figsize=(12, 8))
sns.histplot(data['QTY'], kde=True)
plt.title('Distribution of Purchase Quantities')
plt.show()
```

Purchase quantity is more between 1 to 50

```python
plt.figure(figsize=(12, 8))
sns.histplot(data['VALUE'], kde=True)
plt.title('Distribution of Purchase Values')
plt.show()
```

```python
# Monthly purchasing patterns
monthly_purchase = data.groupby(data['MONTH'])['VALUE'].sum()
monthly_purchase.plot(kind='line', marker='o')
plt.title('Monthly Purchasing Patterns')
plt.xlabel('Month')
plt.ylabel('Total Purchase Value')
plt.show()
```

Here in month 2 there is high purchasing 

```python

from sklearn.cluster import KMeans
# Customer segmentation with larger size
plt.figure(figsize=(12, 8))
kmeans = KMeans(n_clusters=3)
data['cluster'] = kmeans.fit_predict(data[['QTY', 'VALUE']])
sns.scatterplot(data=data, x='QTY', y='VALUE', hue='cluster')
plt.title('Customer Segmentation based on Purchasing Behavior')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
correlation_matrix_idealdata = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_idealdata, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
daily_purchase = Working.groupby(Working['DAY'])['VALUE'].sum()
daily_purchase.plot(kind='line', marker='o')
plt.title('Daily Purchasing Patterns')
plt.xlabel('Day')
plt.ylabel('Total Purchase Value')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=Working, x='MONTH', y='BILL_AMT')
plt.title('Distribution of Bill Amounts Across Months')
plt.xlabel('Month')
plt.ylabel('Bill Amount')
plt.xticks(rotation=0)
plt.show()
```

```python

top_10_brands_idealdata = data.groupby('BRD')['VALUE'].sum().nlargest(10)
top_10_brands_working = Working.groupby('BRD')['BILL_AMT'].sum().nlargest(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_brands_idealdata.index, y=top_10_brands_idealdata.values, label='Ideal Data')
sns.barplot(x=top_10_brands_working.index, y=top_10_brands_working.values, label='Working Data')
plt.xticks(rotation=45)
plt.title('Total Revenue by Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Total Revenue')
plt.legend()
plt.show()

best_brand_idealdata = top_10_brands_idealdata.idxmax()
best_brand_working = top_10_brands_working.idxmax()

print("Best Brand for Profit in Top 10 Brands (Ideal Data):", best_brand_idealdata)
print("Best Brand for Profit in Top 10 Brands (Working Data):", best_brand_working)
```

```python
brand_purchase =Working.groupby('BRD')['BILL_AMT'].sum().sort_values(ascending=False)


top_20_brands = brand_purchase.head(20).index


top_20_data = Working[Working['BRD'].isin(top_20_brands)]


plt.figure(figsize=(12, 8))
sns.boxplot(data=top_20_data, x='BRD', y='BILL_AMT')
plt.title('Distribution of Bill Amounts for Top 20 Brands')
plt.xlabel('Brand')
plt.ylabel('Bill Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
monthly_bill_amount = Working.groupby('MONTH')['BILL_AMT'].sum()

# Plotting
plt.figure(figsize=(12, 8))
sns.lineplot(data=monthly_bill_amount, marker='o', color='blue')
plt.title('Transaction Bill Amount Across Months (Working Dataset)')
plt.xlabel('Month')
plt.ylabel('Total Bill Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
store_bill_amount_ideal = Working.groupby('STORECODE')['BILL_AMT'].sum()

# the store with the highest total bill amount
store_with_highest_bill = store_bill_amount_ideal.idxmax()
highest_bill_amount = store_bill_amount_ideal.max()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=store_bill_amount_ideal.index, y=store_bill_amount_ideal.values, color='skyblue')
plt.title('Bill Amount Across Stores (Working Dataset)')
plt.xlabel('Store Code')
plt.ylabel('Total Bill Amount')
plt.xticks(rotation=45)
plt.axhline(highest_bill_amount, color='red', linestyle='--', label='Store with Highest Bill Amount')
plt.legend()
plt.show()
```

```python
brand_bill_amount = Working.groupby('BRD')['BILL_AMT'].sum()


highest_bill_brand = brand_bill_amount.idxmax()
highest_bill_amount = brand_bill_amount.max()

print("Brand with the highest total bill amount:")
print("Brand:", highest_bill_brand)
print("Total Bill Amount:", highest_bill_amount)
```
